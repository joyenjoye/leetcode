## 常见的双指针方式

1. 同速指针: 链表上有两个指针，一个先出发，另一个后出发并以相同的速度跟随
    - 求链表的逆：通过临时指针让双指针同步前行
    - 求链表的倒数第k个元素：先让其中一个指针向前走k步，接着两个指针以同样的速度一起前进，直到前面的指针走到尽头，则后面的指针即位倒数第k个元素

2. 快慢指针：链表上的两个指针从统一节点出发，其中一个指针前进速度比另个指针快
    - 计算链表的中点：快慢指针从头节点出发，每轮迭代中，快指针向前移动两个节点，慢指针向前移动一个节点，最终当快指针到达终点时，慢指针刚好在中间的节点
    - 判断链表是否有环：快慢指针从头节点出发，如果链表中存在环，两个指针最终会在环中相遇
    - 求链表中环的长度：快慢指针从头节点出发，如果链表中存在环，两个指针最终会在环中相遇。第一次相遇后，第一个指针不动，第二个指针继续前进，计算再次相遇的步数。
